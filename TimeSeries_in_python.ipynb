```python
import pandas as pd
```

```python
temp_NY=pd.read_csv('/Users/akashsoni/Desktop/githubprac/NY_temp.csv')
```

```python
temp_NY.head()
```

```python
temp_NY.set_index('DATE', inplace=True)
```

```python
temp_NY.head()
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
temp_NY.index=pd.to_datetime(temp_NY.index,format='%Y')
```

```python
import matplotlib.pyplot as plt
```

```python
temp_NY.plot()
```

```python
result=adfuller(temp_NY['TAVG'])
```

```python
print("The p-value for the ADF test is",result[1])
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
#Take first difference of the temp series
```

```python
chg_temp=temp_NY.diff()
```

```python
chg_temp=chg_temp.dropna()
```

```python
#plot ACF,PACF on the samepage
```

```python
fig,axes=plt.subplots(2,1)
```

```python
#plot the ACF
```

```python
plot_acf(chg_temp,lags=20,ax=axes[0])
```

```python
#plot the PACF
```

```python
plot_pacf(chg_temp,lags=20,ax=axes[1])
```

```python
#import module for estimate on arma model
```

```python
from statsmodels.tsa.arima_model import ARMA
```

```python
#Fit the data to an AR(1) and print AIC
```

```python
mod=ARMA(chg_temp,order=(1,0))
```

```python
res=mod.fit()
```

```python
print("THE AIC for an AR(1) is :",res.aic)
```

```python
#fit data to an AR(2) modes & print AIC
```

```python
mod=ARMA(chg_temp,order=(2,0))
```

```python
res=mod.fit()
```

```python
print("The AIC is an AR(2) is :",res.aic)
```

```python
#Fit the data to an MA(1) model and print AIC
```

```python
mod=ARMA(chg_temp,order=(0,1))
```

```python
res=mod.fit()
```

```python
print("The AIC for an MA(1) is :",res.aic)
```

```python
#Fit the data to an ARMA(1,1) model and print AIC
```

```python
mod=ARMA(chg_temp,order=(1,1))
```

```python
res=mod.fit()
```

```python
print("The AIC for an ARMA(1,1)is:",res.aic)
```
